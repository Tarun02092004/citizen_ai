```python
import os

# Create project and subfolders
os.makedirs("/content/citizen-ai-engagement-platform/extend", exist_ok=True)
os.makedirs("/content/citizen-ai-engagement-platform/data", exist_ok=True)
```

```python
app_code = """
from transformers import AutoTokenizer, AutoModelForCausalLM
import gradio as gr
import os

token = os.environ.get("HUGGINGFACE_TOKEN")

tokenizer = AutoTokenizer.from_pretrained(
    "ibm-granite/granite-3.3-2b-instruct",
    token=token
)

model = AutoModelForCausalLM.from_pretrained(
    "ibm-granite/granite-3.3-2b-instruct",
    token=token
)

def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        do_sample=True,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

iface = gr.Interface(
    fn=generate_response,
    inputs=gr.Textbox(lines=5, placeholder="Type your query here", label="Citizen Query"),
    outputs=gr.Textbox(lines=10, label="AI Response"),
    title="Citizen AI Engagement Platform"
)

iface.launch(share=True)
"""

with open("/content/citizen-ai-engagement-platform/app.py", "w") as f:
    f.write(app_code)
```
