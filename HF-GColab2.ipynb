```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import gradio as gr
import torch

# Model name on Hugging Face
model_name = "ibm-granite/granite-3.3-2b-instruct"

# Load tokenizer and model
print("Loading tokenizer and model...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
print("Model loaded!")
```

```python
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=512,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
def chat_with_citizen_ai(prompt):
    return generate_response(prompt)

iface = gr.Interface(
    fn=chat_with_citizen_ai,
    inputs=gr.Textbox(lines=5, placeholder="Ask about city services, policies, or report issues here...", label="Citizen Query"),
    outputs=gr.Textbox(lines=10, label="AI Response"),
    title="Citizen AI Engagement Platform",
    description="Interact with the AI to get information or submit feedback about civic issues."
)
```

```python
iface.launch(share=True)
```
